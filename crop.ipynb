```python
features = crop.columns.to_list()
features.remove("label")
print(features)
```

```python
# Visualization of distribution of each feature
fig, ax = plt.subplots(3, 3, figsize = (15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.histplot(crop[i], ax = subplot, kde = True)
    subplot.set_title(i)
plt.tight_layout()
plt.show()
```

```python
# Plotting scatter plots
fig, ax = plt.subplots(3, 3, figsize=(15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.scatterplot(x=i, y="label", data=crop, ax=subplot)
    subplot.set_title(i)
plt.tight_layout()
plt.show()
```

```python
# Finding outliers - boxplot
fig, ax = plt.subplots(3, 3, figsize = (15, 15))
for i, subplot in zip(features, ax.flatten()):
    sns.boxplot(crop[i], ax = subplot)
plt.tight_layout()
plt.show()
```

```python
num_cols = crop.select_dtypes(include=[np.number])
num_cols
```

```python
corr = num_cols.corr()
corr
```
